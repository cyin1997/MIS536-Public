### Select rows using loc and iloc 
```loc``` is label-based, which means that you have to specify rows and columns based on their row and column labels. ```iloc``` is integer index based, so you have to specify rows and columns by their integer index like you did in the previous exercise.


````python
housing_df.loc[0:3]
````

````python
housing_df.iloc[0:4]
````

````python
housing_df.loc[10:20]
````

````python
housing_df.iloc[10:21]
````

````python
housing_df.loc[-10:] # note that this doesn't work !
````

````python
housing_df.iloc[-10:] # starts at index 10 before last and goes until end
````

````python
housing_df.iloc[:-10] # starts at the first record ad stops 10 before last 
````

## Select Columns


### Renaming columns

The names of columns can often be long, not descriptive, or inconvenient to write. Columns names can also have hidden space. 


````python
housing_df.columns
````

To rename a column, you can either explicitly rename a single column, or you could write Python code to search for patterns in names and programatically rename a set of columns. 

#### Explicit renaming of a column

````python
housing_df = housing_df.rename(columns={'TOTAL VALUE ': 'TOTAL_VALUE'})# explicit
````

#### Dynamically (programatically) renaming multiple columns


````python
housing_df.columns = [s.strip().replace(' ','_') for s in housing_df.columns] # list comprehension, all columns
````

### Select Single Column


#### Select by column name (be careful! this results in returning series object, not a dataframe)


````python
housing_df["TOTAL_VALUE"] # this results in a series, not a dataFrame!
````

````python
housing_df[["TOTAL_VALUE"]] # this results in a dataframe
````

#### Selecting columns using loc and iloc


##### Using loc


````python
housing_df.loc[:, 'TOTAL_VALUE']
````

````python
housing_df.loc[:, ['TOTAL_VALUE']]
````

````python
housing_df.loc[:,['TOTAL_VALUE', 'TAX']] # the columns here are in a list
````

##### Using iloc

````python
housing_df.iloc[:, 0]
````

````python
housing_df.iloc[:, [0]]
````

````python
housing_df.iloc[:, 0:2]
````

````python
housing_df.iloc[:, [0,2,10]] #  Note that the column indeces 0, 2 and 10 are placed in a list
````
